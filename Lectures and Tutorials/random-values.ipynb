### Setup

***

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')
plt.rcParams["figure.figsize"] = 10,8
```

```python
#set to None to generate random numbers
seed = 1
```

```python
# Create an insance of a random number generator
rng = np.random.default_rng(seed=seed)
```

<br>

### Uniform Distribution

***

```python
# Generate single uniform value
rng.uniform()
```

```python
# Generate single value between 10 and 100
rng.uniform(low=10.0,high=100.0)
```

```python
# Generate several numbers
nos = rng.uniform(size=10000)
```

```python
plt.hist(nos);
```

<br>

### Normal Distribution

***

```python
#single value
rng.normal()
```

```python
# generate 10000 values
nos = rng.normal(size=10000)
```

```python
plt.hist(nos);
```

```python
sns.kdeplot(nos,shade=True,color='blue');
```

```python
#generate two datasets
nos_A = rng.normal(loc=1.0, size=10000)
nos_B = rng.normal(loc=3.0, size=10000)
```

```python
fig, ax = plt.subplots()

ax.hist(nos_A,bins=100)
ax.hist(nos_B,bins=100);
```

```python
sns.kdeplot(nos_A,shade=True)
sns.kdeplot(nos_B,shade=True);
```

```python
#generate two datasets
nos_A = rng.normal(loc=5.0,scale=0.1, size=10000)
nos_B = rng.normal(loc=5.0,scale=0.5, size=10000)
```

```python
fig, ax = plt.subplots()

ax.hist(nos_A,bins=40,alpha=.5,density=True)
ax.hist(nos_B,bins=40,alpha=.5,density=True);
```

### Binomial Distribution

***

```python
#single number
rng.binomial(1,p=0.5)
```

```python
#single number
nos = rng.binomial(1,p=0.5,size=10000)
```

```python
#create barchart
plt.bar(['Fail','Success'], [nos.size - nos.sum(),nos.sum()]);
```

### Normal Samples

```python
#generate 10,000 samples of normal geerated samples
samples = rng.normal(loc=0.0,size=(10000,100))
```

```python
len(samples[0])
```

```python
samples.shape
```

```python
plt.hist(samples.flatten(),bins=20);
```

```python
#calculate mean of all numbers
samples.mean()
```

```python
#calculate mean across samples
mean_samples = samples.mean(axis=1)
mean_samples
```

```python
plt.hist(mean_samples,bins=20);
```

```python
samples_A = rng.normal(loc=1.0,scale=10.0,size=(10000,100))
samples_B = rng.normal(loc=1.1,scale=10.0,size=(10000,100))
```

```python
#calculate means of samples
mean_samples_A = samples_A.mean(axis=1)
mean_samples_B = samples_B.mean(axis=1)
```

```python
plt.hist(mean_samples_A,bins=40,alpha=0.5,density=True)
plt.hist(mean_samples_B,bins=40,alpha=0.5,density=True);
```
