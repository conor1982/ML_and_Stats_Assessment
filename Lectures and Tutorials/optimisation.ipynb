**Square root of 2**

When $x^2 = 2$ what is $x$?

$X^2 = 2 \Rightarrow x^2 - 2=0$

$ f(x) = x^2 - 2 $

When $f(x) = 0$ what is $x?$

***

```python
# function above
def f(x):
    return x**2 - 2.0
```

```python
#x's
x = np.linspace(-3.0,3.0,1000)
```

```python
fig, ax = plt.subplots()

#x axis plot
ax.plot(x,0*x)

ax.plot(x,f(x));
```

#### Derivitive

polynomial $f(x) = x^2 - 2$

derivitive $f^\prime(x) = 2x - 0 = 2x$

$y = mx+c =2x + 0$
***

```python
def df(x):
    return 2.0 * x
```

```python
fig, ax = plt.subplots()

#x axis plot
ax.plot(x,0*x)

#plot f(x)
ax.plot(x,f(x));

#plot f'(x)
ax.plot(x,df(x));
```

```python
#The slope of f(x) when x is 3.0
df(3.0)
```

```python
#The value of f(x) when x = 3.0
f(3.0)
```

$y - y_1 = m(x-x_1)$

$y = m(x-x_1)+y_1$


***

```python
# tangent of f(x) at f(3.0)
tan = (df(3.0)*(x-3.0)) + f(3)
```

```python
fig, ax = plt.subplots()

#x axis plot
ax.plot(x,0*x)

#plot f(x)
ax.plot(x,f(x));

#plot f'(x)
ax.plot(x,df(x))

#plot tan
ax.plot(x,tan,linewidth=2);
```

```python
#generic tangent calculation
def tan(x,x1):
    return (df(x1)*(x-x1)) + f(x1)
```

```python
fig, ax = plt.subplots()

#x axis plot
ax.plot(x,0*x)

#plot f(x)
ax.plot(x,f(x));

#plot f'(x)
ax.plot(x,df(x))

#plot tan at 1.0
ax.plot(x,tan(x,1.0));
```

#### Integral

* opposite of the derivitive

$f(x) = x^2 - 2 = x^2 - 2 x^0$

$F(x) = \frac{x^3}{3} - 2 \frac{x^1}{1} = \frac{1}{3}x^3 -2x$, ignoring the constant
 
$F(x) = \frac{x^3}{3} - 2 \frac{x^1}{1} = \frac{1}{3}x^3 -2x + c $ with constant


***

```python
#intregal in python
def F(x):
    return ((x**3)/3.0) - (2.0*x)
```

```python
fig, ax = plt.subplots()

#x axis plot
ax.plot(x,0*x)

#plot f(x)
ax.plot(x,f(x));

#plot f'(x)
#ax.plot(x,df(x))

#plot tan at 1.0
ax.plot(x,F(x));
```

#### Gradient Descent

***

```python
#require accuaracy
accuracy = 0.01
```

```python
# Initial Guess
guess = 3.0

#learning rate
learning_rate = 0.01

#iterations
i = 0

# slowly decrease x and watch the value of F

while abs(f(guess)) > accuracy:
    #print(f"X: {guess:4.3f} F(x): {F(guess):6.3f}")
    if f(guess) < 0.0:
        guess = guess + learning_rate
    else:
        guess = guess - learning_rate
        
    i = i+1
    if i >= 1000:
        print("Max Iterastions Reached")
        break

print("")
print(f"FINAL RESULTS X: {guess:4.3f} F(x): {F(guess):6.3f}")
```

```python
def sqrt_Func_Newton(x):

    #Approx start point dividing input number by 2
    approx = x/2
        
    #Loop while absolute of value of f(x) is greater or equal than 0.01
    #Ref https://www.youtube.com/watch?v=2158QbsunA8
    while abs(approx**2-x) >= 0.01:
        approx = approx - (approx**2 - x)/(2*approx)

    #Return approx number at end of loop rounded to 1 decimel place
    return round(approx,9)
```

```python
sqrt_Func_Newton(2)
```
